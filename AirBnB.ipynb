```python
columns_to_keep = ['id', 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
columns_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license', 'id']
```

```python
len(columns_to_keep)
type(columns_to_keep)
```

```python
len(columns_to_drop)
```

```python
#Method for dropping columns
data.columns = [col.lower() for col in data.columns]
data = data[[col for col in data.columns if col in columns_to_keep]]
columns_to_drop_lower = [col.lower() for col in columns_to_drop]
data.columns = [col.lower() for col in data.columns]
columns_to_drop_existing = [col for col in columns_to_drop_lower if col in data.columns]
data.drop(columns=columns_to_drop_existing, inplace=True)

columns_to_keep_lower = [col.lower() for col in columns_to_keep]
data_columns_lower = [col.lower() for col in data.columns]
selected_columns = [data.columns[data_columns_lower.index(col)] for col in columns_to_keep_lower if col in data_columns_lower]
data = data[selected_columns]
data.columns
```
